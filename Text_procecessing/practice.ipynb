### Romove html tags

```python
import re

html_count = 0
def Remove_html(text):
  global html_count
  pattern = re.compile(r'<.*?>') 
  if pattern.search(text):
    html_count += 1
  return pattern.sub(" ", text) # substitute / replace
```

```python
df['review'] = df['review'].apply(Remove_html)
print(html_count)
```

```python
df.head()
```

### Remove URL

```python
count = 0
def Remove_URL(text):
  global count
  parttern = re.compile(r"https?://\S+|www\.\S+")
  if parttern.search(text):
    count += 1
  return parttern.sub(" ", text)
```

```python
df['review'] = df['review'].apply(Remove_URL)
print(count)
```

### Puntuation Remove

```python
pun_count = 0
def Remove_punctuation(text):
  global pun_count
  pattern = re.compile(r'[^\w\s]')
  if pattern.search(text):
    pun_count += 1
  return pattern.sub(" ", text)
```

```python
df['review'] = df['review'].apply(Remove_punctuation)
print(pun_count)
```

```python
chat_words_count = 0
def replase_chat_word(text, chat_words):
  global chat_words_count 
  new_text = []
  spilt_text = text.split()

  for word in spilt_text:
    if word.upper() in chat_words:
      chat_words_count += 1
      new_text.append(chat_words[word.upper()])
    else:
      new_text.append(word)
  return " ".join(new_text)
```

```python
df['review'] = df['review'].apply(lambda x : replase_chat_word(x, chat_words))
print(chat_words_count)
```

### Stop word remove

```python
from nltk.corpus import stopwords
```

```python
stop_word = stopwords.words('english')
```

```python
def remove_stopwords(text):
  new_text = []

  for word in text.split():
    if word in stop_word:
      new_text.append('')
    else:
      new_text.append(word)
  
  return " ".join(new_text)
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

### Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()

def stem_word(text):

  split_word = text.split()

  ps_word = []
  for word in split_word:
    ps_word.append(ps.stem(word))  # from token to string
  
  return " ".join(ps_word)
```

```python
df['review'] = df['review'].apply(stem_word)
```

### Lammatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
Wordnet_lemmatizer = WordNetLemmatizer()
def lemmatizer(text):
  sen_split = text.split()

  lamma_word = []

  for word in sen_split:
    lamma_word.append(Wordnet_lemmatizer.lemmatize(word, pos = "v"))
  
  return lamma_word
```

```python
lemmatizer("He is playing football")
```

```python
df['review'] = df['review'].apply(lemmatizer)
```
